```python
df['id'] = (df['imageurl']).astype('category').cat.codes  
```

```python
app = ClarifaiApp(api_key='e2971b53cb274c05a018154c0d6cc4f6')
model = app.models.get('color')

data_final = pd.DataFrame()

for id in df['id']:
    url_img = df.loc[df['id']==id, 'imageurl'].iloc[0]
    image = ClImage(url=str(url_img))
    response = model.predict([image])
    for i in range(0,len(response['outputs'][0]['data']['colors']),1):
        p={'id': id, 'color': str(response['outputs'][0]['data']['colors'][i]['raw_hex']),'value': response['outputs'][0]['data']['colors'][i]['value']}
        data_final = data_final.append(p, ignore_index=True)
```
